```ipython kernel install --user --name=numan```

If you had a numan kernel installed before , or you are unsure if yo uhave it, look at the list of available kernels.
```jupyter kernelspec list```
You can then delete the old one if it exists:
```jupyter kernelspec uninstall numan```
And create a new one ( inside the new numan enviroment):
```ipython kernel install --user --name=numan```
